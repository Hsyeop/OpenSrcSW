```python
import csv

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
max_d_list = [-999, 0, 999, 0, -999, 0]
header = next(data)
for d in data:
    if d[0] == '':
        break
    if max(float(d[4]) - float(d[3]),
           max_d_list[0]) == float(d[4]) - float(d[3]):
        max_d_list[0], max_d_list[1] = float(d[4]) - float(
            d[3]), d[0]
    if min(float(d[3]), max_d_list[2]) == float(d[3]):
        max_d_list[2], max_d_list[3] = float(d[3]), d[0]
    if max(float(d[4]), max_d_list[4]) == float(d[4]):
        max_d_list[4], max_d_list[5] = float(d[4]), d[0]
f.close()
print('일교차가 가장 큰 날')
print(max_d_list[1])
print('최저기온이 가장 낮은 날')
print(max_d_list[3])
print('최고기온이 가장 높은 날')
print(max_d_list[5])
```

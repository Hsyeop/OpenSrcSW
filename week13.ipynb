```python
import matplotlib.pyplot as plt
import csv
f = open('tenYears.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
h_temp , l_temp =[], []
for d in list(data):
    if '10-28' in d[0]:
        l_temp.append(float(d[3]))
        h_temp.append(float(d[4]))

    
plt.title('2010~2020 1028 highest, lowest temperature')
plt.plot(h_temp,'pink')
plt.plot(l_temp,'skyblue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('tenYears.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan,aug = [],[]
for d in list(data):
    if '-01-' in d[0]:
        jan.append(float(d[-1]))
    if '-08-' in d[0]:
        aug.append(float(d[-1]))
    
plt.title('2020-01, 2020-08 highest temperature histogram')
plt.hist(aug,bins=100,color='red',label='Aug')
plt.hist(jan,bins=100,color='blue',label='Jan')
plt.legend()
plt.show()
```
